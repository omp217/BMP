```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the csv file latency.csv and convert it to a dataframe and first row as header

df = pd.read_csv('latency.csv', header=0)
df = df.reindex(df.mean().sort_values(ascending=False).index, axis=1)
df.head()
```

```python
columns = df.columns

# Plot the single graph for all the columns 

plt.figure(figsize=(10,10))

for column in columns:
    plt.plot(df[column], label=column + ' - Mean Value: ' + str(round(df[column].mean(), 2)), linewidth=0.7)
    plt.legend()

plt.xlabel('Iteration')
plt.ylabel('Time (ms)')
plt.title('Latency Graph')
plt.show()
```
