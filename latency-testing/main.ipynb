```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the csv file latency.csv and convert it to a dataframe with columns name of rmvvrm and mvvm

df = pd.read_csv('latency.csv', names=['rmvvrm', 'mvvm'])
df.head()
```

```python
# plot one graph in which on the y axis print time of two methods which are columns of dataframe and on the x axis print the number of iterations

plt.plot(df['rmvvrm'], label='rmvvrm')
plt.plot(df['mvvm'], label='mvvm')
plt.xlabel('Iterations')
plt.ylabel('Time')
plt.legend()
```

```python
# Print average time of each method

print('Average time of rmvvrm: ', df['rmvvrm'].mean())
print('Average time of mvvm: ', df['mvvm'].mean())
```
